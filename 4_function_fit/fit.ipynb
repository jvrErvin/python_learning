```python
from scipy.optimize import curve_fit # Az illesztéshez használt függvény, python -m pip install scipy
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0, 1, 20)           # É.T.
zaj = 0.3 * np.random.randn(len(x)) # normál eloszlású zaj

a = 2
b = 1
y_lin_noisy = (a * x + b) + zaj # É.K.
```

```python
plt.plot(x, y_lin_noisy, 'o', label='mért adatok zajjal')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.ylim(0,5)
plt.grid(True)

plt.vlines(1.1, 0, 5, color='r', linestyle='--')

plt.show()
```

Lineáris függvény (egyenes):
$$ f(x) = a \cdot x + b $$

```python
def lin(x, a, b):
    return a * x + b
```

```python
popt, pcov = curve_fit(lin, x, y_lin_noisy) # az illesztés elvégzése
perr       = np.sqrt(np.diag(pcov))         # az illesztési paraméterek hibáinak meghatározása

print(f"a = {popt[0]} +- {perr[0]}")
print(f"b = {popt[1]} +- {perr[1]}")
```

```python
x_for_fitted = np.linspace(0, 2, 30)

plt.plot(x_for_fitted, lin(x_for_fitted, popt[0], popt[1]), color='red',linewidth=3, label="illesztett függvény")
plt.plot(x, y_lin_noisy, 'o', label='mért adatok')

plt.vlines(1.1, 0, 5, color='r', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
