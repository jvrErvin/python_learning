```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
```

----
# 1. Basics, ábrázolás Jupyter varázslattal

```python
# Értelmezési tartomány
x = np.linspace(0, 2*np.pi, 100)

# Érték készlet
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# Canvas létrehozás
plt.figure(figsize=(10, 6), facecolor="white")

# Ábrázolás: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x, y_sin, marker="o", linestyle="--", color="purple", label="sin(x)") # marker méret, vonal vastagság, ....
plt.plot(x, y_cos, "rx--", label="cos(x)")

# Tengelyfeliratok
plt.xlabel("x", fontsize = 20)
plt.ylabel("y", fontsize = 20)

# Ábracím
plt.title("sin(x) és cos(x)", fontsize = 20)

# Rácsok
plt.grid(visible=True, axis="both") # axis=both,x,y

# X tengely feliratok elforgatása
plt.xticks(rotation=90)

# X tengely feliratok lecserélése
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ["0", "pi/2", "pi", "3pi/2", "kutya"])
#plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])

# Jelmagyarázat
plt.legend(loc = "best", fontsize = 20) # loc="upper right", "lower right", "upper left", "lower left", "best"

# Ábra mentése
plt.savefig("./sin_cos.png")

plt.show()
```

Minden ábrára muszály:
- tengelyfeliratok
- cím
- legend

----
# 2. Ugyan ez csak rendesen

```python
# Ábra létrehozása
fig, ax = plt.subplots(figsize=(10, 6), facecolor="white")

# Ábrázolás
ax.plot(x, y_sin, marker="o", linestyle="dashed", color="purple", label="sin(x)")
ax.plot(x, y_cos, "rx", label="cos(x)")

# Tengelyfeliratok
ax.set_xlabel("x", fontsize = 20)
ax.set_ylabel("y", fontsize = 20)

# Ábracím
ax.set_title("sin(x) és cos(x)", fontsize = 20)

# Rácsok
ax.grid(visible=True, axis="both") # axis=both,x,y

# X tengely feliratok megadása és elforgatása
ax.xaxis.set_major_locator(FixedLocator([0, np.pi/2, 2*np.pi]))
ax.set_xticklabels(["0", "pi/2", "2pi"], rotation=45)

# Jelmagyarázat
ax.legend(loc="upper right", fontsize = 20)

plt.show()
```

----
# 3. Subplot

```python
# 2x2 ábra
fig, axs = plt.subplots(2, 2, figsize=(10, 6), facecolor="white", sharey=True, sharex=True) # sharey=True: azonos y tengely, sharex=True: azonos x tengely

axs[0, 0].plot(x, y_sin, "g", label="sin(x)")
axs[0, 1].plot(x, y_cos, "b", label="cos(x)")

axs[1, 0].plot(x, y_cos, "r", label="cos(x)")
axs[1][1].plot(x, y_sin, "y", label="sin(x)") # indexelés: [sor, oszlop] vagy [sor][oszlop]

plt.show()
```
