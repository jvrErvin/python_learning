```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
```

```python
x = np.linspace(0, 2*np.pi, 200)

y_sin = np.sin(x)
y_sin2 = np.sin(x * 2)
y_sin0_5 = np.sin(x * 0.5)
y0_5_sin = 0.5 * np.sin(x)
y2_sin = 2 * np.sin(x)
```

```python
plt.figure(figsize = (10, 6), facecolor="white")

plt.plot(x, y_sin, marker = "<", linestyle = "-", color = "purple", label = "sin(x)")
plt.plot(x, y_sin2, marker = "<", linestyle = "-", color = "red", label = "sin(2x)")
plt.plot(x, y_sin0_5, marker = "<", linestyle = "-", color = "green", label = "sin(0.5x)")
plt.plot(x, y0_5_sin, marker = "<", linestyle = "-", color = "magenta", label = "0.5sin(x)")
plt.plot(x, y2_sin, marker = "<", linestyle = "-", color = "aqua", label = "2sin(x)")

plt.title("sin(x) és többszörösei", fontsize = 20)
plt.grid(visible = True, axis = "both")
plt.xlabel("x", fontsize = 40)
plt.ylabel("y", fontsize = 40)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ["0", "pi/2", "pi", "3pi/2", "2pi"], fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend()

plt.show()
```
