```python
#pip install scikit-learn
from sklearn.linear_model import LinearRegression

import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 1, 20)           # É.T.
zaj = 0.3 * np.random.randn(len(x)) # normál eloszlású zaj

a = 2
b = 1
y_lin_noisy = (a * x + b) + zaj # É.K.
```

$$ y = f(x) = a \cdot x + b $$

- a: egységnyi változás az x-ben mennyivel változtatja meg f(x) függvény értékét
- b: az f(x) függvény értéke, amikor x = 0

```python
model = LinearRegression()
```

```python
x.shape
```

```python
x.reshape(-1, 1).shape
```

```python
model.coef_, model.intercept_
```

```python
model.fit(x.reshape(-1, 1), y_lin_noisy)
```

```python
a_fitted = model.coef_[0]
b_fitted = model.intercept_

print("a = ", a_fitted)
print("b = ", b_fitted)
```

```python
plt.plot(x, y_lin_noisy, 'o', label='mért adatok zajjal')

x_foo = np.linspace(-1, 2, 100)
plt.plot(x_foo, a_fitted * x_foo + b_fitted, '-', label='illesztett egyenes')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.ylim(0,5)
plt.grid(True)


plt.show()
```
